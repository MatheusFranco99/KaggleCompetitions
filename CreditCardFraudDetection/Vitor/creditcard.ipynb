```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('creditcard.csv')

data.head()
```

```python
data.info()
```

```python
# checking for missing values
data.isnull().sum()
```

```python
# plotting Non Fraud (0) and Fraud (1) number of transactions
count_class = data['Class'].value_counts()

count_class.plot(kind='bar', rot=0)

plt.title('Non Fraud x Fraud')
plt.xlabel("Class")
plt.ylabel("Number of transactions")

for i, value in enumerate(count_class):
    plt.text(i-0.05, value, str(value), fontsize=10)

plt.show()
```

```python
# valeu thunder. checking variable correlations
plt.figure(figsize=(16, 16))
sns.heatmap(data.corr(),cmap='RdYlBu',annot=True,fmt=".1f",linewidth=.5)
plt.show()
```

```python
# fraud/ valid transactions ratio -> unbalanced dataset
valid = data[data['Class']==0]
fraud = data[data['Class']==1]
fraud_valid_ratio = round((len(fraud)/ len(valid)) * 100, 4)

print(f"Ratio: {fraud_valid_ratio}% fraud detections")
```
