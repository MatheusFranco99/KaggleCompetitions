```python
# PROFILING


# histogram

import matplotlib.pyplot as plt
import math
import numpy as np

def numberOfBins(values,method = 'Sturges'):

    n = len(values)

    if n <= 1:
        return 1

    if method == 'Sturges':
        return int(math.log2(n) + 1)
    elif method == 'Sqrt':
        return int(math.sqrt(n))
    elif method == 'Scott':
        return max(int((3.5 * np.std(values)) / (pow(n,1/3))),1)
    else:
        return 100


def classificationHistograms(df,classifications = [0,1], bins = 100, class_column_name = 'Class', bins_rule = 'Sqrt'):
    column_names = df.columns

    n = len(column_names)
    m = len(classifications)

    fig, ax = plt.subplots(n-1,m,figsize=(m*8,(n-1)*6))

    for i,column_name in enumerate(column_names):
        
        if column_name != class_column_name:

            for j,classification in enumerate(classifications):
                values = df[df[class_column_name] == classification][column_name]
                n_bins = numberOfBins(values, method=bins_rule)
                ax[i,j].hist(values, bins=n_bins)
                ax[i,j].set_title(f'Histogram of {column_name}, class={classification}, bins = {n_bins}')
                ax[i,j].set_xlabel(f'{column_name}')
                ax[i,j].set_ylabel('Frequency')

    plt.show()

classificationHistograms(df, classifications=[0,1], bins=100, class_column_name = 'Class')
```
